```python
df.dtypes

# Checking data type. No Need to change.
```

```python
nulls = df.isna().sum()

nulls
# Checking for Null Values.
```

```python
df.loc[df['Country Label'].isnull()]
```

```python
# Define a dictionary mapping country codes to country labels
country_mapping = {
    'XK': 'Kosovo',
    'KV': 'Kosovo',
    'UC': 'Curacao',
    'CW': 'Curacao',
    'AQ': 'American Samoa',
    'AJ': 'Azerbaian',
    'BK': 'Bosnia and Herzegovina',
    'CE': 'Sri Lanka',
    'IZ': 'Iraq',
    'KU': 'Kuwait',
    'MK': 'North Macedonia',
    'TI': 'Tajikistan',
    'TX': 'Turkmenistan',
    'VM': 'Vietnam'
}

# Replace country codes with country labels using the dictionary
df['Country Label'] = df['Country Code'].replace(country_mapping)


# This is necessary because some countries are missing values in the 'country label' column, through some research, we can figure this out.
```

```python
# List of columns to check for missing values... there are some missing ("NaN") values in these columns.
columns_to_fill = ['City', 'Location', 'Coordinates']

# Fills missing values in specified columns with 'Missing Data'
for column in columns_to_fill:
    df[column] = df[column].fillna('Missing data')

# Counts missing values for each column
nulls = {column: df[column].isna().sum() for column in df.columns}
nulls
```

```python
print("Before dropping duplicates:", df.shape) # count before the dropping of duplicate rows
df.drop_duplicates(inplace=True) # Drops any duplicate rows
print("After dropping duplicates:", df.shape) # count after the dropping of duplicate rows
```

```python
# Dropping rows with 'Value' less than 0
df = df.drop(df[df['Value'] < 0].index)

# Dropping rows with certain 'Unit' values
units_to_drop = ['c', 'particles/cm³', '%']
df = df.drop(df[df['Unit'].isin(units_to_drop)].index)
```
